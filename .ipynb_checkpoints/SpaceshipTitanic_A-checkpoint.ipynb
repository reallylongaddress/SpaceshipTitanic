# Build a preprocessing processing pipeline

```python
# # --BREAK APART string into components to see if there's any interesting information hidden in there....
# Cabin            object (unique 6561, total 8693)(null: 199)
# PassengerId      object (null: 0)

# class parseCabin(BaseEstimator, TransformerMixin):

#   def __init__(self):
#     pass

#   def transform(self,X,y=None):
        
#     # print(f'shape: {X.shape}::{type(X)}::{X.columns}')
#     cabin_A = ''
#     cabin_B = ''
#     cabin_C = ''

#     for index, row in X.iterrows():
#         if (row["Cabin"] is not np.nan) and (len(row["Cabin"].split('/')) == 3):

#             cabin_parts = row["Cabin"].split('/')

#             cabin_A = cabin_parts[0]
#             cabin_B = cabin_parts[1]
#             cabin_C = cabin_parts[2]

#         X.at[index, 'Cabin_A'] = cabin_A
#         X.at[index, 'Cabin_B'] = cabin_B
#         X.at[index, 'Cabin_C'] = cabin_C
        
#     self.columns = X.columns
    
#     # print(f'self.columns: {self.columns}')
#     print(f'parseCabin.transform: {type(X)}::{X.shape}::{X.columns}')
#     return X

#   def fit(self, X, y=None):
#     return self   

#   def get_feature_names_out(self, arg):
#     return self.columns

##---------------------

class parsePassengerId(BaseEstimator, TransformerMixin):

  def __init__(self):
    pass

  def transform(self,X,y=None):
        
    # print(f'shape: {X.shape}::{type(X)}::{X.columns}')
    passenger_id_A = ''
    passenger_id_B = ''

    for index, row in X.iterrows():

        passenger_parts = row["PassengerId"].split('_')

        passenger_id_A = passenger_parts[0]
        passenger_id_B = passenger_parts[1]

        X.at[index, 'PassengerId_A'] = passenger_id_A
        X.at[index, 'PassengerId_B'] = passenger_id_B
        
    self.columns = X.columns
    
    # print(f'self.columns: {self.columns}')
    # print(f'parsePassengerId.transform: {type(X)}::{X.shape}::{X.columns}')
    return X

  def fit(self, X, y=None):
    return self   

  def get_feature_names_out(self, arg):
    return self.columns

##---------------------

class parseAndImputeCabins(BaseEstimator, TransformerMixin):

  def __init__(self):
    pass

  def transform(self,X,y=None):

    #parse the cabin into 3 parts
    for index, row in X.iterrows():
        cabin_A = np.nan
        cabin_B = np.nan
        cabin_C = np.nan

        if (row["Cabin"] is not np.nan) and (len(row["Cabin"].split('/')) == 3):

            cabin_parts = row["Cabin"].split('/')

            cabin_A = cabin_parts[0]
            cabin_B = cabin_parts[1]
            cabin_C = cabin_parts[2]
            
        X.at[index, 'Cabin_A'] = cabin_A
        X.at[index, 'Cabin_B'] = cabin_B
        X.at[index, 'Cabin_C'] = cabin_C
        
    self.columns = X.columns
    
    #impute the cabin parts
    cabin_mf_imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
    X[['Cabin_A', 'Cabin_C']] = cabin_mf_imputer.fit_transform(X[['Cabin_A', 'Cabin_C']])
    
    cabin_mean_imputer = SimpleImputer(strategy='mean', missing_values=np.nan)
    X[['Cabin_B']] = cabin_mean_imputer.fit_transform(X[['Cabin_B']])
    X['Cabin_B'] = X['Cabin_B'].astype(int)

    #one hot encode Cabin_A & Cabin_C

    return X
    
  def fit(self, X, y=None):
    return self   

  def get_feature_names_out(self, arg):
    return self.columns


##---------------------

class cabin_oh_encoder(BaseEstimator, TransformerMixin):

  def __init__(self):
    pass

  def transform(self,X,y=None):
    print("-------hello")
    return X
    
  def fit(self, X, y=None):
    return self   

  def get_feature_names_out(self, arg):
    return self.columns

##---------------------

class planetImputerEncoder(BaseEstimator, TransformerMixin):

  def __init__(self):
    pass

  def transform(self,X,y=None):
    print("-------planetImputerEncoder hello")
    
    imputer = SimpleImputer(strategy='most_frequent')
    imputer = imputer.fit(X[['HomePlanet']])
    X['HomePlanet'] = imputer.transform(X[['HomePlanet']])#.ravel()
    
    imputer = imputer.fit(X[['Destination']])
    X['Destination'] = imputer.transform(X[['Destination']])#.ravel()
    
    
    #one hot encode HomePlanet & Destination
    
    self.columns = X.columns
    
    return X
    
  def fit(self, X, y=None):
    return self   

  def get_feature_names_out(self, arg):
    return self.columns
##---------------------

class cryoVIPEncoder(BaseEstimator, TransformerMixin):

  def __init__(self):
    pass

  def transform(self,X,y=None):
    print("-------cryoVIPEncoder hello")
    
    #impute with most frequent (True/False) then convert to (1/0)
    imputer = SimpleImputer(strategy='most_frequent')
    imputer = imputer.fit(X[['CryoSleep']])
    X['CryoSleep'] = imputer.transform(X[['CryoSleep']])
    X['CryoSleep_bool'] = X['CryoSleep'].map({True:1, False:0})
    
    
    imputer = imputer.fit(X[['VIP']])
    X['VIP'] = imputer.transform(X[['VIP']])
    X['VIP_bool'] = X['VIP'].map({True:1, False:0})
    
    self.columns = X.columns
    
    return X
    
  def fit(self, X, y=None):
    return self   

  def get_feature_names_out(self, arg):
    return self.columns

##---------------------

class TrueFalseEncoder(BaseEstimator, TransformerMixin):

  def __init__(self):
    pass

  def transform(self,X,y=None):
    for column in X.columns:
        X[f'{column}_bool'] = X[column].map({True:1, False:0})
    self.columns = X.columns
    
    return X

  def fit(self, X, y=None):
    return self   

  def get_feature_names_out(self, arg):
    print(self.columns)
    return self.columns

##---------------------

class columnDropperTransformer(BaseEstimator, TransformerMixin):

  def __init__(self,drop_columns):
    self.drop_columns = drop_columns
    print(f'columnDropperTransformer: {self.drop_columns}')
    self.return_columns = None

  def transform(self,X,y=None):
    print(f'transform drop: {self.drop_columns}')
    print(f'cols: {X.columns}')
    rv = X.drop(self.drop_columns, axis=1)
    self.return_columns = rv.columns
    return rv

  def fit(self, X, y=None):
    return self 

  def get_feature_names_out(self, arg):
    return self.return_columns

##---------------------
```

```python
_train = train_df.copy()
print(f'_train: {_train.shape}, {_train.columns}')
imputer_ct = ColumnTransformer(
    [
        ('parse_passenger_id', parsePassengerId(), ['PassengerId']),
        ('imput_zero', SimpleImputer(strategy='constant', fill_value=0), ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']),
        ('imp_mean', SimpleImputer(strategy='mean'), ['Age']),
        ('planet_transformer', planetImputerEncoder(), ['HomePlanet', 'Destination']),
        ('cryo_vip_transformer', cryoVIPEncoder(), ['CryoSleep', 'VIP']),
        ('imp_most_freq_B', SimpleImputer(strategy='most_frequent'), ['CryoSleep', 'VIP']),
        ('encoder_true_false', TrueFalseEncoder(), ['Transported']), 
        ('parse_and_impute_cabins', parseAndImputeCabins(), ['Cabin']),
    ],
    remainder='passthrough',
    verbose_feature_names_out=False
)

```

```python
cols_to_drop = ['passengerId', 'CryoSleep', 'Transported', 'Cabin', 'Name']


oh_ct = ColumnTransformer(
    [
        ('encoder_oh', OneHotEncoder(sparse_output=False, drop='if_binary'), ['Cabin_A', 'Cabin_C', 'HomePlanet', 'Destination']),
    ],
    remainder='passthrough',
    verbose_feature_names_out=False
)

p = Pipeline(
    [
        ('imputer_ct', imputer_ct),
        ('oh_ct', oh_ct),
        ('column_dropper', columnDropperTransformer(cols_to_drop)),
    ]
)

_train = train_df.copy()

z = p.fit_transform(_train)
z
```
